```python
#import libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#Load the csv file (world_population dataset) into Jupyter Notebooks using the read_csv method providing the file path of dataset's location locally in your pc as the method function's parameter
df = pd.read_csv(r"C:\Users\Martina\Desktop\Owen\Exploratory Data Analysis\world_population.csv")
df
```

```python
#Transfor the float values to have 2 decimal values from one decimal value using the set_option method
pd.set_option('display.float_format', lambda x: '%.2f' % x)
```

```python
#Run the data drame to confirm that the changes have been effected. df is variable that stores the dataset's information
df
```

```python
# Using the info method, we're able to have a quick overview of our dataset information such as the dataset's columns, number of unique values(Not-Null) in each column, and the data type of values stores in the rows of each column(int64(stands for numeric values), object(stands for string/ alphanumeric character values), and float64 (stands for numeric values that are accopanied by numeric values after a decimal point))
df.info()
```

```python
# In order to get the general overview of the descriptive statistics of our dataset we run the describe method. The describe method will display the count, mean, standard deviation(std), minimum value(min), percentiles(25%,50%,75%) and the maximum value(max) of ou dataset
df.describe()
```

```python
#the isnull method accompanied by the sum method helps us to know the sum of null values for each column in the dataset
df.isnull().sum()
```

```python
#the nunique method helps us to know the sum of non-null(unique) values for each column in the dataset

df.nunique()
```

```python
#the sort_values method enables us to sort our dataset based on a pre-defined columneg(2022 Population) and criteria of sorting (ascending(ascending=True) or descending(ascending=False) as the parameter values, the head method helps us to specifythat we need to filter out our result set on the top ten values.
df.sort_values(by="2022 Population", ascending=False).head(10)
```

```python
# df.columns helps us to see all of our dataset's columns
df.columns
```

```python
#The groupby method helps us aggregate our data based on the 'Continent' by their mean values as specified by the mean method
df2 = df.groupby('Continent')[['1970 Population',
       '1980 Population', '1990 Population', '2000 Population',
       '2010 Population', '2015 Population', '2020 Population',
       '2022 Population']].mean().sort_values(by="2022 Population", ascending=False)
df2
```

```python
# The transpose method effectively interchanges rows and columns whereby rows become columns and vice versa
df3 = df2.transpose()
df3
```

```python
#The plot method plotsaline graph showing the population growth by continent over the years from 1970 t0 2022
df3.plot(figsize=(20,10))
```

```python
# The boxplot method plots a boxplot visualization that effectively enables us spot outliers by how unevenly a data point is distributed as compared to oher data points.
df.boxplot(figsize=(20,10))
```

```python
#The corr method is used to show how correlated the datapoints are whereby 1 represents high correlation and below 0.5 represents little to no correlation
df.corr()
```

```python
sns.heatmap(df.corr(), annot=True)

plt.rcParams['figure.figsize'] = (20,7)

plt.show()
```
